```python
import requests
#specifying the url
wiki_link="https://en.wikipedia.org/wiki/Maruti_Suzuki"
#converting url as text
link=requests.get(wiki_link).text
```

```python
link
```

```python
#performing parse
from bs4 import BeautifulSoup
soup=BeautifulSoup(link,'lxml')
soup
```

```python
#Structuring the data
soup.prettify()
```

```python
soup.title.string
```

```python
right_tables=soup.find('table',class_="wikitable")
right_tables
```

```python
table_links=right_tables.findAll('a',class_="mw-redirect")
table_links
```

```python
car=[]
production=[]
for links in table_links:
    
  
        car.append(links.get('title'))

print(car)
```

```python
import pandas as pd
df=pd.DataFrame()
df['Carname']=car
df
```
